```python
x = 10 # Global variable
def update_x():
    global x # declared that we are using the global variable x
    x += 10 # modifay the global variable 
update_x()
print(x)
```

```python
x = 10 # global variable 
def update_x():
    globals()['x'] += 20 # Access and modifay the global variable using the dictionary
    
update_x()
print(x)
```

How to pass list  to a function 

```python
def count(list):
    even = 0
    odd = 0
    
    for i in list :
        if i%2==0:
            even += 1
        else :
            odd += 1
    return even,odd
lst = [10,9,34,5,6,425,12,100]
even,odd = count(lst)
print(even)
print(odd)
```

```python
def count(list):
    even = 0
    odd = 0
    
    for i in list :
        if i%2==0:
            even += 1
        else :
            odd += 1
    return even,odd
lst = [10,9,34,5,6,425,12,100]
even,odd = count(lst)
print('Even Numbers :{} and odd numbers :{}'.format(even,odd))

```
