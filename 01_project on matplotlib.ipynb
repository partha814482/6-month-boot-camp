```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python


# line plot
x = np.linspace(0,10,100)
y = np.sin(x)

plt.figure(figsize=(8,4))
plt.plot(x, y, label = "sin(x)", color = 'blue')
plt.title("Sine Function")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.legend() # Don't forget to add this line to display the label
plt.show()
```

This code generates and displays a line plot of the sine function using the `NumPy` and `Matplotlib` libraries in Python.

Here is a breakdown of what each part of the code does:

* **`import numpy as np` and `import matplotlib.pyplot as plt`**: These lines import the necessary libraries. `NumPy` is used for numerical operations, specifically for creating arrays of numbers, and `Matplotlib.pyplot` is used for plotting.

* **`x = np.linspace(0, 10, 100)`**: This line creates a NumPy array named `x`. `np.linspace()` generates 100 evenly spaced values between 0 and 10 (inclusive). These values will serve as the x-coordinates for the plot.

* **`y = np.sin(x)`**: This line calculates the sine of each value in the `x` array. The result is a new array `y` where each element is the sine of the corresponding element in `x`. These values will be the y-coordinates.

* **`plt.figure(figsize=(8, 4))`**: This line creates a new figure (the window or canvas where the plot will be drawn). The `figsize` argument sets the dimensions of the figure to be 8 inches wide and 4 inches tall.

* **`plt.plot(x, y, label="sin(x)", color='blue')`**: This is the core plotting command.
    * `plt.plot()` draws a line graph.
    * `x` and `y` are the data points to be plotted.
    * `label="sin(x)"` assigns a label to this line. This label will be used in the legend to identify the plot.
    * `color='blue'` sets the color of the line to blue.

* **`plt.title("Sine Function")`**: This adds a title to the plot, which appears at the top.

* **`plt.xlabel("x-axis")`**: This labels the x-axis.

* **`plt.ylabel("y-axis")`**: This labels the y-axis.

* **`plt.grid(True)`**: This adds a grid to the plot, making it easier to read the values.

* **`plt.legend()`**: This displays the legend, which shows the labels of each plotted line (in this case, "sin(x)"). This is a crucial step to make the `label` argument in `plt.plot()` visible.

* **`plt.show()`**: This command displays the final plot on the screen. It is necessary to render the figure and all the elements that have been added to it.

```python
#pie chat
labels = ['Sleep','work','Expercise','Leisure','Others']
sizes = [8,9,2,3,2]
colors =['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0']

plt.figure(figsize=(6,6))
plt.pie(sizes, labels = labels,colors=colors,autopct = '%1.1f%%', startangle = 140)
plt.title('Daily Time Allocation')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle

plt.show()
```

This code generates and displays a pie chart using the `Matplotlib` library in Python. The chart visualizes the daily allocation of time for various activities.

Here is a breakdown of what each part of the code does:

* **`labels = ['Sleep','work','Expercise','Leisure','Others']`**: This line creates a list of strings that will be used as the labels for each slice of the pie chart.

* **`sizes = [8,9,2,3,2]`**: This line creates a list of numerical values. Each number represents the "size" or proportion of a corresponding slice in the pie chart. In this case, it represents the number of hours spent on each activity.

* **`colors =['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0']`**: This line creates a list of hexadecimal color codes. Each code specifies the color for each slice of the pie chart, corresponding to the `labels` and `sizes` lists.

* **`plt.figure(figsize=(6,6))`**: This line creates a new figure (the canvas for the plot) with a specific size. `figsize=(6,6)` sets the figure's dimensions to be 6 inches by 6 inches, which is useful for creating a square plot for the pie chart.

* **`plt.pie(sizes, labels = labels, colors=colors, autopct = '%1.1f%%', startangle = 140)`**: This is the main command for drawing the pie chart.
    * `plt.pie()` is the function used to create the pie chart.
    * `sizes`: This argument provides the numerical data that determines the size of each pie slice.
    * `labels = labels`: This argument assigns the text labels to each slice.
    * `colors=colors`: This argument assigns the custom colors to the slices.
    * `autopct = '%1.1f%%'`: This argument controls the format of the percentage values that are displayed on each slice. The string `'%1.1f%%'` formats the percentage to one decimal place, followed by a percentage sign.
    * `startangle = 140`: This argument rotates the start of the pie chart. The default is 0 degrees, but here it's rotated by 140 degrees counter-clockwise from the x-axis, which can improve the visual layout of the chart.

* **`plt.title('Daily Time Allocation')`**: This adds a title to the pie chart, providing context for the data.

* **`plt.axis('equal')`**: This command ensures that the pie chart is drawn as a perfect circle instead of an ellipse. It does this by adjusting the aspect ratio of the plot so that the scaling of the x-axis and y-axis is the same.

* **`plt.show()`**: This final command displays the complete pie chart with all its elements (slices, labels, title) in a new window.

```python
# Histogarm
scores = np.random.normal(loc = 70, scale = 10,size = 1000)

plt.figure(figsize=(8,4))
plt.hist(scores,bins=30,color='skyblue',edgecolor = 'black')
plt.title('Exam score Distribution')
plt.xlabel('score')
plt.ylabel('Number of student')
plt.grid(True)
plt.show()
```

 *scores = np.random.normal(loc=70, scale=10, size=1000): This line generates a NumPy array named scores.

==np.random.normal() generates random numbers from a normal (Gaussian) distribution.

==loc=70: Sets the mean (average) of the distribution to 70.

==scale=10: Sets the standard deviation of the distribution to 10. This controls the spread of the data.

==size=1000: Specifies that 1000 random numbers should be generated.

==The result is an array of 1000 simulated exam scores, with most of the scores clustered around 70.



```python
# Heatmap
data = np.random.rand(10,10)
plt.figure(figsize=(6,5))
plt.imshow(data,cmap = 'hot',interpolation='nearest')
plt.colorbar(label = 'Value')
plt.title('Random Heatmap')
plt.show()
```

This code generates and displays a heatmap using the `NumPy` and `Matplotlib` libraries. A heatmap is a graphical representation of data where the individual values contained in a matrix are represented as colors.

Here is a breakdown of what each part of the code does:

* **`import numpy as np` and `import matplotlib.pyplot as plt`**: These lines import the necessary libraries for numerical operations (`NumPy`) and plotting (`Matplotlib`).

* **`data = np.random.rand(10, 10)`**: This line creates a 10x10 NumPy array of random numbers.
    * `np.random.rand(10, 10)` generates random floating-point numbers from a uniform distribution between 0 and 1.
    * This array, `data`, will be the input for the heatmap, where each number's value will be mapped to a specific color.

* **`plt.figure(figsize=(6, 5))`**: This line creates a new figure (the plot window) with a width of 6 inches and a height of 5 inches.

* **`plt.imshow(data, cmap='hot', interpolation='nearest')`**: This is the main command for generating the heatmap.
    * `plt.imshow()` is the function used to display data as an image (in this case, a heatmap).
    * `data`: This is the 2D array that provides the values to be color-mapped.
    * `cmap='hot'`: This argument specifies the colormap to use. The `'hot'` colormap is a popular choice for heatmaps, where lower values are mapped to dark colors and higher values are mapped to bright, hot colors like yellow and white.
    * `interpolation='nearest'`: This argument controls the interpolation method used to display the image. `'nearest'` means that each data point (each cell in the 2D array) will be displayed as a single, solid-colored pixel, without any smoothing between pixels. This is a good choice for showing discrete data.

* **`plt.colorbar(label='Value')`**: This adds a color bar next to the heatmap.
    * A color bar is a key that maps the colors in the heatmap to their corresponding data values.
    * `label='Value'` sets the title of the color bar.

* **`plt.title('Random Heatmap')`**: This adds a title to the plot.

* **`plt.show()`**: This final command displays the complete heatmap on the screen in a new window.
