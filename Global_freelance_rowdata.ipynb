```python
clean_data['is_active']
```

```python
clean_data['is_active'] = clean_data['is_active'].astype(str).str.upper().replace({'0': 'No','1': 'Yes','TRUE': 'Yes','FALSE': 'No','N': 'No','Y': 'Yes'})
```

.astype(str)

Converts all values in the is_active column to strings.

This ensures consistency, because the column may contain a mix of int (0,1), bool (True, False), or string values ('Y', 'N').

.str.upper()

Converts everything to uppercase.

Example: 'true' → 'TRUE', 'y' → 'Y'.

This makes replacement easier because you don’t have to worry about case variations.

.replace({...})

Uses a dictionary to map different possible values to consistent values:

'0' → 'No'

'1' → 'Yes'

'TRUE' → 'Yes'

'FALSE' → 'No'

'N' → 'No'

'Y' → 'Yes'

```python
clean_data['client_satisfaction']
```

```python
clean_data['client_satisfaction'] = (
    clean_data['client_satisfaction']
    .astype(str)                         # make sure values are strings
    .str.replace('%', '', regex=False)   # remove the '%' symbol
)

clean_data['client_satisfaction'] = pd.to_numeric(clean_data['client_satisfaction'], errors='coerce')
```

```python
clean_data['client_satisfaction'].fillna(
    clean_data['client_satisfaction'].mean(),
    inplace=True
)
```

```python
clean_data.isnull().sum()
```

```python
clean_data
```

```python
clean_data['hourly_rate (USD)']
```

```python
# Remove unwanted text/symbole and keep only numbers
clean_data['hourly_rate (USD)'] = (clean_data['hourly_rate (USD)'].astype(str).str.replace(r'[^\d.]','',regex=True))
# convert to numeric
clean_data['hourly_rate (USD)'] = pd.to_numeric(clean_data['hourly_rate (USD)'],errors='coerce')
# .astype(str)  ==ensure string
#  .str.replace(r"[^\d.]", "", regex=True)  remove non-numeric characters
```

errors="coerce" → If a value is invalid, pandas converts it into NaN (Not a Number, the standard missing value in pandas).
